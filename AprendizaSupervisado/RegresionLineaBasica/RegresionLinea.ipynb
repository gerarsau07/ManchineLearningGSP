# Exporando Molelos con Regresion Lineal


```python
import pandas as pd

dt = pd.read_csv("grados.csv")
```

```python
dt.info()
```

```python
dt.head()
```

```python
import seaborn as sb

sb.scatterplot(
    x="celsius", # dato para la variable x
    y="fahrenheit", # dato para la variable x
    data=dt,
    hue="fahrenheit" #color depedinedo del dato
)
```

```python
# Separar datos

x = dt["celsius"]
y = dt["fahrenheit"]
```

```python
X = x.values.reshape(-1,1)
Y = y.values.reshape(-1,1)
```

```python
x , y
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
modelo.fit(X,Y)
```

```python
modelo.predict([[1123]])
```

```python
modelo.score(X,Y)
```
