```python
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer(as_frame=True)
```

```python
cancer.frame.head()
```

```python
cancer.frame.info()
```

```python
cancer.data.shape
```

```python
from sklearn.model_selection import train_test_split

x = cancer.data
y = cancer.target

x_ent, x_pru, y_ent, y_pru = train_test_split(x,y, test_size=.2)
```

```python
from sklearn import svm
from sklearn import metrics

kernels = ["linear", "rbf", "sigmoid"]
gammas = [1, .01, .001, .0001, .000001]

for kernel in  kernels:
    for gamma in gammas:
        modelo = svm.SVC(kernel=kernel, gamma=gamma)
        modelo.fit(x_ent, y_ent)
        predicciones = modelo.predict(x_pru)
        print(f"{kernel} :, Gamma: {gamma} Exactitud: ", metrics.accuracy_score(y_pru, predicciones))
        print(f"{kernel} :, Gamma: {gamma} Exactitud: ", metrics.precision_score(y_pru, predicciones))
```

```python
modelo = svm.SVC(kernel="linear")
modelo.fit(x_ent, y_ent)
predicciones = modelo.predict(x_pru)
print(f"{kernel} :, Gamma: {gamma} Exactitud: ", metrics.accuracy_score(y_pru, predicciones))
print(f"{kernel} :, Gamma: {gamma} Exactitud: ", metrics.precision_score(y_pru, predicciones))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_pru, predicciones))
```

```python
from sklearn.metrics import confusion_matrix

pd.DataFrame(confusion_matrix(y_pru, predicciones))
```
